## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import ipywidgets as widgets
from ipywidgets import interact, fixed
from tsfresh import extract_features
from tsfresh.feature_extraction import EfficientFCParameters
from tsfresh import select_features

plt.rcParams["figure.figsize"] = (14, 4)
pd.options.display.max_columns = 500
```

## Load

```python
daily_train = pd.read_csv("../data/processed/daily-train-base.csv", parse_dates=["ds"])
daily_train
```

## Features

```python
ys = daily_train[["id", "ds", "y"]].copy()
```

```python
daily_train_features = extract_features(
    ys, 
    column_id="id", 
    column_sort="ds", 
    default_fc_parameters=EfficientFCParameters(),
)
```

```python
daily_train_features.to_csv("../data/processed/daily-train-features.csv", index=False)
```
